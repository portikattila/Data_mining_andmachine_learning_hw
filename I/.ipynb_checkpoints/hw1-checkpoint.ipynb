If they have, investigate if the missingness is random or not.<a href="http://patbaa.web.elte.hu/physdm/code_examples/01_SOLVED_EDA.html#1.-Load-the-above-linked-csv-file-as-a-pandas-dataframe.-Check-&amp;-plot-if-any-of-the-columns-has-missing-values.-If-they-have,-investigate-if-the-missingness-is-random-or-not." class="anchor-link">¶</a>

Impute the missing values in a sensible way:

-   if only a very small percentage is missing, imputing with the
    column-wise mean makes sense, or also removing the missing rows
    makes sense
-   if in a row almost all the entries is missing, it worth to remove
    that given row
-   if a larger portion is missing from a column, usually it worth to
    encode that with a value that does not appear in the dataset (eg:
    -1).

The imputing method affects different machine learning models different
way, but now we are interested only in EDA, so try to keep as much
information as possible!

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import sklearn as sk
from scipy.stats import pearsonr
import seaborn as sns
```

## Taks 1

```python
url="http://patbaa.web.elte.hu/physdm/data/titanic.csv"
titaninc_dataset=pd.read_csv(url)
titaninc_dataset.head()
```

```python
titaninc_dataset.isna().sum()
```

```python
missing=titaninc_dataset.isna()


fig, ax = plt.subplots(figsize=(24,2))
ax.imshow(missing.values.sum(axis=0).reshape((12,1)).T, cmap=cm.viridis, vmax=1000)


#major ticks
ax.set_xticks(np.arange(0, 12, 1))
ax.set_yticks([])

#Labels for major ticks
ax.set_xticklabels(titaninc_dataset.columns.values)

#minor ticks
ax.set_xticks(np.arange(-0.5, 12.5, 1), minor=True)
ax.set_yticks(np.arange(-1.5, 1, 1), minor=True)
ax.grid(which='minor', color='k', linestyle='-', linewidth=4)

ax.set_xlim(-0.52,11.51)
ax.set_ylim(-0.52,0.52)

ax.tick_params(labelsize=15)

ax.set_title('Missing values in columns', fontsize=25)

for i,x in enumerate(np.arange(0, 12, 1)):
    ax.text(x-0.1,-0.1,str(missing.values.sum(axis=0).reshape((12,1)).T[0][i]),fontsize=20,color='w')
plt.show()
```

```python
#encode the missing values in the 'Cabin' column with 'N/A'
titaninc_dataset.Cabin=titaninc_dataset.Cabin.fillna('N/A')

#encode the missing values in the 'Age' column with -1
titaninc_dataset.Age=titaninc_dataset.Age.fillna(-1)

#delete the rows which do not contain a value in the last column.
titaninc_dataset=titaninc_dataset.drop(index=titaninc_dataset.index[missing.values[:,-1]])
```

```python
missing=titaninc_dataset.isna()


fig, ax = plt.subplots(figsize=(24,2))
ax.imshow(missing.values.sum(axis=0).reshape((12,1)).T, cmap=cm.viridis, vmax=1000)


#major ticks
ax.set_xticks(np.arange(0, 12, 1))
ax.set_yticks([])

#Labels for major ticks
ax.set_xticklabels(titaninc_dataset.columns.values)

#minor ticks
ax.set_xticks(np.arange(-0.5, 12.5, 1), minor=True)
ax.set_yticks(np.arange(-1.5, 1, 1), minor=True)
ax.grid(which='minor', color='k', linestyle='-', linewidth=4)

ax.set_xlim(-0.52,11.51)
ax.set_ylim(-0.52,0.52)

ax.tick_params(labelsize=15)

ax.set_title('Missing values in columns', fontsize=25)

for i,x in enumerate(np.arange(0, 12, 1)):
    ax.text(x-0.1,-0.1,str(missing.values.sum(axis=0).reshape((12,1)).T[0][i]),fontsize=20,color='w')
plt.show()
```

```python
#the pairwise correlation of columns 
CORR=np.abs(titaninc_dataset.corr())
fig,ax=plt.subplots(figsize=(10,9))
sns.heatmap(CORR, cmap=cm.viridis)

plt.show()
```

```python
fig,ax=plt.subplots(figsize=(9,6))
h1,_=np.histogram(titaninc_dataset.Pclass[titaninc_dataset.Cabin.values=='N/A'],bins=3, density=True)
h2,_=np.histogram(titaninc_dataset.Pclass, bins=3, density=True)

ax.bar([0.95,1.95,2.95],h1, color='tab:blue', alpha=0.9,label='missing cabin number')
ax.bar([1.05,2.05,3.05],h2, color='tab:orange',alpha=0.9, label='all passenger')
ax.set_xticks([1,2,3])
ax.set_yticks([0,0.5,1])

ax.set_xlabel('Pclass', fontsize=20)
ax.set_ylabel('Number of passengers', fontsize=20)
ax.legend()
plt.show()

fig,ax=plt.subplots(figsize=(9,6))
h1,_=np.histogram(titaninc_dataset.Pclass[titaninc_dataset.Cabin.values=='N/A'],bins=3)
h2,_=np.histogram(titaninc_dataset.Pclass, bins=3)


ax.bar([1,2,3],h2, color='tab:blue', label='all passenger')
ax.bar([1,2,3],h1, color='tab:red',label='missing cabin number')
ax.set_xticks([1,2,3])
ax.set_yticks(h1)

ax.set_xlabel('Pclass', fontsize=20)
ax.set_ylabel('Number of passengers', fontsize=20)
ax.legend()
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(9,6))
h1,_=np.histogram(titaninc_dataset.Pclass[titaninc_dataset.Age.values==-1],bins=3, density=True)
h2,_=np.histogram(titaninc_dataset.Pclass, bins=3, density=True)

ax.bar([0.95,1.95,2.95],h1, color='tab:blue', alpha=0.9,label='unknonw age')
ax.bar([1.05,2.05,3.05],h2, color='tab:orange',alpha=0.9, label='all passenger')
ax.set_xticks([1,2,3])
ax.set_yticks([0,0.5,1])

ax.set_xlabel('Pclass', fontsize=20)
ax.set_ylabel('Number of passengers', fontsize=20)
ax.legend()
plt.show()

fig,ax=plt.subplots(figsize=(9,6))
h1,_=np.histogram(titaninc_dataset.Pclass[titaninc_dataset.Age.values==-1],bins=3)
h2,_=np.histogram(titaninc_dataset.Pclass, bins=3)


ax.bar([1,2,3],h2, color='tab:blue', label='all passenger')
ax.bar([1,2,3],h1, color='tab:red',label='unknonw age')
ax.set_xticks([1,2,3])
ax.set_yticks(h1)

ax.set_xlabel('Pclass', fontsize=20)
ax.set_ylabel('Number of passengers', fontsize=20)
ax.legend()
plt.show()
```

```python
#the pairwise correlation of the missing cabin number and Pclass
cabin=np.copy(titaninc_dataset.Cabin.values)
cabin[cabin=='N/A']=1.0
cabin[cabin!=1]=0.0
cabin=np.asarray(cabin, dtype=float)
Pclass=np.copy(titaninc_dataset.Pclass.values)
class_cabin=pd.DataFrame(data=np.array((Pclass, cabin)).T, columns=['Pclass', 'Cabin'])
Pclass=np.asarray(Pclass, dtype=float)
corr,_ = pearsonr(Pclass, cabin)
print('Pearsons correlation: %.3f' % corr)
```

<font size=4>
The cabin number missing systematically from the dataframe. For passengers from the lower classes, the cabin number was incompletely recorded.

<font size=4>
According to the correlation matrix, the surviving is correlated to the passenger class and with the fare.  The relation of the class and fare is obvious: the higher classes are more expensive. The reason that wealthier passengers had more chances to survive the catastrophe is also clear. They counted more important than the poor passengers, even in the last minutes. Another practical reason could be that cabins of the higher passenger classes are in safe locations on the boat, as long as the cabins of the poorer passengers were placed near the bottom of the ship.

```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111)
ax.set_title('Sex ratio of passengers', fontsize=30)
age=np.copy(titaninc_dataset.Age.values)
ax.hist(age, bins=20, density=True, alpha=0.8)
age=np.copy(titaninc_dataset.Age.values[titaninc_dataset.Survived.values==1])
ax.hist(age, bins=20, density=True, alpha=0.8)
ax.tick_params(labelsize=20)

plt.show()
```

<font size=4>

Based on that, one can expect that the age distribution of the passengers is shifted, and the average age of the survivors are less than the average age of the passengers who entered the ship, but the dataset shows a different result. The average age and the distribution remain roughly the same.
