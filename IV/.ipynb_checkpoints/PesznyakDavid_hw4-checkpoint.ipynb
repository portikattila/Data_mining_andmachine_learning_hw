```python
# importing used packages/libraries/methods
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

# Task 1.
### Load the provided .npy files. You can load them with numpy.

* Each file contains one vector, $X$ and $y$.
* Visualize $X$ vs $y$ on a scatter plot.
* Fit an  $y=w_0 + w_1⋅X + w_2⋅X^2$  linear regression using `sklearn`.

```python
# load files
X = np.load('X.npy')
Y = np.load('y.npy')
```

```python
# scatter plot of X and y
plt.figure(figsize = (15, 10))
plt.plot(X, Y, 'o', alpha = 0.5)
plt.grid()
plt.xlabel('X', size = 20)
plt.ylabel('Y', size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
# linear regression with sklearn
# format data
dataX = np.array([X, X ** 2]).T
# define model and fit data
reg = LinearRegression().fit(dataX, Y)
predY = reg.predict(dataX)
```

```python
# check model
# scatter plot of X and Y
plt.figure(figsize = (15, 10))
plt.scatter(X, Y, alpha = 0.5, label = 'data')
plt.scatter(X, predY, alpha = 0.5, label = 'prediction')
plt.grid()
plt.xlabel('$X$', size = 20)
plt.ylabel('$Y$', size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.legend(fontsize = 20)
```

```python
# get coefficients
print('Intercept: ', reg.intercept_)
print('Coefficients: ', reg.coef_)
```

## Comments
The fitted coefficents are $w_0\approx-47.37$, $w_1\approx1.19$ and $w_2\approx0.06$.

# Task 2.
### Using different features

* Plot the residuals (the difference between the prediction and the actual  $y$) vs the original  $y$.
* A non-random-noise like pattern suggests non-linear connection between the features and the predictions.
* Someone told us that the connection between $X$ and $y$ is  $y=A⋅X+B⋅\cos^3(X)+C⋅X^2+D$.
  * Using sklearn's linear regression estimate  $A$,$B$,$C$,$D$!
* Plot the residuals again! is it better now?

```python
# plotting residuals
plt.figure(figsize = (15, 10))
plt.scatter(Y, Y - predY, alpha = 0.5)
plt.grid()
plt.xlabel('$Y$', size = 20)
plt.ylabel('$Y-Y_{pred}$', size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
# format data
dataX_new = np.array([X, np.cos(X) ** 3, X ** 2]).T
# predicting parameters of the new model
reg_new = LinearRegression().fit(dataX_new, Y)
predY_new = reg_new.predict(dataX_new)
```

```python
# get coefficients
print('Intercept: ', reg_new.intercept_)
print('Coefficients: ', reg_new.coef_)
```

```python
# check new model
# scatter plot of X and Y
plt.figure(figsize = (15, 10))
plt.scatter(X, Y, alpha = 0.5, label = 'data')
plt.scatter(X, predY_new, alpha = 0.5, label = 'prediction')
plt.grid()
plt.xlabel('$X$', size = 20)
plt.ylabel('$Y$', size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.legend(fontsize = 20)
```

```python
# plotting new residuals
plt.figure(figsize = (15, 10))
plt.scatter(Y, Y - predY_new, alpha = 0.5)
plt.grid()
plt.xlabel('$Y$', size = 20)
plt.ylabel('$Y-Y_{pred}$', size = 20)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

## Comments
The residuals are much better: smaller in magnitude, seemingly no ordered structure.

# Task 3.
### Other methods than sklearn for linear regression

* Using the `statsmodels` package perform the same linear regression as in 2.) (hint: `use statsmodels.api.OLS`).
* Is the result the same? If not guess, why? (Did you not forget to add the constant term?).
* Try to get the same results with statsmodels as with sklearn!
* Using the analytic solution formula shown during the lecture, calculate the coefficients ($A$, $B$, $C$, $D$). Are they the same compared to the two previous methods?

```python
# linear regression with statsmodels
# add constant
dataX_sm = dataX_new
dataX_sm = sm.add_constant(dataX_sm)
# fit the model
model = sm.OLS(Y, dataX_sm)
```

```python
# coefficients from statsmodels
results = model.fit()
print(results.summary())
```

```python
# coefficients from sklearn
print('SKLEARN: ')
print('Intercept: ', reg_new.intercept_)
print('Coefficients: ', reg_new.coef_)
```

```python
# analytic formula for the coefficients
w = np.linalg.inv(dataX_sm.T @ dataX_sm) @ dataX_sm.T @ Y
print('Analytic oefficients: ', w)
```

## Comments
The results are the same with `sklearn`, `statsmodels` and the analytic formula.

# Task 4.

* Load the [real_estate](https://gist.github.com/qbeer/f356d7144543cbb09c9792c34b8ad722) data to a pandas dataframe
drop the ID column and the geographic location columns
fit a linear regression model to predict the unit price using sklearn.

```python
# load data
dataRE = pd.read_csv('real_estate.csv')
# head of data
dataRE.head()
```

```python
# drop ID and geographic location data
dataRE = dataRE.drop(['No', 'X5 latitude', 'X6 longitude'], axis = 1)
# data to estimate
YRE = dataRE.pop('Y house price of unit area')
```

```python
# fit model
regRE = LinearRegression().fit(dataRE, YRE)
```

```python
# coefficients from sklearn
print('Intercept: ', regRE.intercept_)
print('Coefficients: ', regRE.coef_)
```
