### 1. Load the provided .npy files. You can load it with numpy.

* each file contains one vector, X and y
* visualize X vs y on a scatter plot
* fit an  y=w_0 + w_1⋅X + w_2⋅X^2  linear regression using `sklearn`
-------------

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
#load files
X = np.load('X.npy')
Y = np.load('y.npy')
```

```python
fig,ax=plt.subplots(1,1, figsize=(8,8))
ax.plot(X, Y, 'o', alpha = 0.5)
ax.grid()
ax.set_xlabel('X', size = 20)
ax.set_ylabel('Y', size = 20)
plt.show()
```

```python
#define data
data=np.array([X, X ** 2]).T
#define model
reg=LinearRegression().fit(data, Y) 

Y_pred=reg.predict(data)
```

```python
fig,ax=plt.subplots(1,1, figsize=(8,8))
ax.plot(X, Y, 'o', alpha = 0.5, label='data')
ax.plot(X, Y_pred, '.',label='predicted values')
ax.grid()
ax.set_xlabel('X', size = 20)
ax.set_ylabel('Y', size = 20)
ax.legend()
plt.show()
```

```python
# get coefficients
print(' w_0 ', reg.intercept_,'\n w_1: ', reg.coef_[0],'\n w_2 :',reg.coef_[1])
```

-------------------
### 2. Using different features

* plot the residuals (the difference between the prediction and the actual  y ) vs the original  y 
* a non-random-noise like pattern suggests non-linear connection between the features and the predictions
* someone told us that the connection between X and y is  y=A⋅X+B⋅cos^3(X)+C⋅X^2+D 
  * using sklearn's linear regression estimate  A,B,C,D !
* plot the residuals again! is it better now?
------------------

```python
fig,ax=plt.subplots(1,1, figsize=(8,8))
ax.plot(Y, Y-Y_pred, 'o', alpha = 0.5)
ax.grid()
ax.set_xlabel('Y', size = 20)
ax.set_ylabel('Y-Y$_{pred}$', size = 20)
plt.show()
```

```python
# format data
data=np.array([X, np.cos(X) ** 3, X ** 2]).T

reg=LinearRegression().fit(data, Y)
Y_pred= reg.predict(data)
```

```python
# get coefficients
print(' A ', reg.intercept_,
      '\n B: ', reg.coef_[0],
      '\n C :',reg.coef_[1],
      '\n D :',reg.coef_[1])
```

```python
fig,ax=plt.subplots(1,1, figsize=(8,8))
ax.plot(X, Y, 'o', alpha = 0.5, label='data')
ax.plot(X, Y_pred, '.',label='predicted values')
ax.grid()
ax.set_xlabel('X', size = 20)
ax.set_ylabel('Y', size = 20)
ax.legend()
plt.show()
```

```python
fig,ax=plt.subplots(1,1, figsize=(8,8))
ax.plot(Y, Y-Y_pred, 'o', alpha = 0.5)
ax.grid()
ax.set_xlabel('Y', size = 20)
ax.set_ylabel('Y-Y$_{pred}$', size = 20)
plt.show()
```

The residual is one order of magnitude smaller, and it looks incoherent.

### 3. Other methdods than sklearn for linear regression

* using the statsmodels package perform the same linear regression as in 2.) (hint: use statsmodels.api.OLS)
* is the result the same? if not guess, why? (did you not forget to add the constant term?)
* try to get the same results with statsmodels as with sklearn!
* using the analytic solution formula shown during the lecture, calculate the coefficients (A, B, C, D). are they the same compared to the two previous methods?
--------------

```python
#regression with statsmodels
data=sm.add_constant(data)
#fit the model
model=sm.OLS(Y, data)
```

```python
# coefficients from statsmodels
result=model.fit()
# get coefficients
print(' A ',result.params[0],
      '\n B: ',result.params[1],
      '\n C :',result.params[2],
      '\n D :',result.params[3])
```

```python
#coefficients from the analytical formula 
coeff = np.linalg.inv(data.T @ data) @ data.T @ Y
print(' A ',coeff[0],
      '\n B: ',coeff[1],
      '\n C :',coeff[2],
      '\n D :',coeff[3])
```

The coefficients from the three methods are roughly the same.

### 4.

* load the [real_estate](https://gist.github.com/qbeer/f356d7144543cbb09c9792c34b8ad722) data to a pandas dataframe
drop the ID column and the geographic location columns
fit a linear regression model to predict the unit price using sklearn
-------------------


```python
data=pd.read_csv('real_estate.csv')
data.head(10)
```

```python
# drop columns
data=data.drop(['No', 'X5 latitude', 'X6 longitude'], axis = 1)
Y = data.pop('Y house price of unit area')
```

```python
reg=LinearRegression().fit(data, Y)
```

```python
print('w_0 =', reg.intercept_)
for i in range(len(reg.coef_)):
    print('w_'+str(i+1),'=',reg.coef_[i])
```

### 5.

* interpret the coefficients and their meaning shortly with your own words
* plot the residuals for the predictions. if you had to decide only on this information, which house would you buy?

---

```python
Y_pred=reg.predict(data)
fig,ax=plt.subplots(1,1, figsize=(8,8))
ax.plot(Y, Y-Y_pred, 'o', alpha = 0.5)
ax.grid()
ax.set_xlabel('Y', size = 20)
ax.set_ylabel('Y-Y$_{pred}$', size = 20)
plt.show()
```

I would like to buy a house near to zero on the vertical axis because there is the real value of the house and the price of it is close to each other. 
If the aim of the buying is making a profit, then one would buy a house with a large deviation between the predicted and the real value. Because in that case, it is possible to sell at a higher price than its original price.

* interpretations:
    * $X_1$ transaction data: is a date upon which a trade takes place, the value of a house depends on the status of real estate market
    * $X_2$ the age of house: the older the houses usually are cheaper
    * $X_3$ distance to the nearest MRT station: houses too far or too close to stations are cheaper.
    * $X_4$ number of convenience stores: houses that have around more stores are more expensive.
