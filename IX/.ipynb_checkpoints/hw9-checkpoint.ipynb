### 1. Implement a logistic regression model on MNIST for binary classification in Numpy

* use `from sklearn.datasets import load_digits`
  * the provided data files: X_train, X_test are a split generated by the author of this exercise
  * w, b are the fitted logistic regression weight and bias, the predictions on the provided test set can be generated by: `y_pred_test = sigmoid(np.dot(w, X_test.T) + b)`
  * the binarized accuracy score should be above 98% for this particular data
* divide the data into 0 vs. 1-9. categories
  * the goal is to separate the zeros
* y = logistic(wx + b)
* optimize the model via stochastic mini-batch gradient descent until you reach above 90% accuracy on the test set


```python
#impoirt libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
def sigmoid(x):
    """sigmoid function"""
    return 1/(1 + np.exp(-x))

def logistic_loss(y_pred,y):
    """Logistic loss function"""
    return -y * np.log(y_pred) - (1-y)*np.log(1-y_pred)

#the gradients of the logistic loss
def logistic_loss_db(y_pred,y):
    return np.sum((y_pred-y))
    
def logistic_loss_dw(X,y_pred,y):
    return np.sum(((y_pred-y)*X),axis=0).reshape((1,X.shape[1]))

#create prediction
def predict(X,w,b):
    return (sigmoid(np.dot(w, X.T) + b)).reshape((X.shape[0],1))

#accuracy score
def ACC(y_pred,y):
    return np.sum(y==np.round(y_pred)) / y.size

#calculate and plot the confusion matrix
def conf_matrix(y,y_pred, figsize=(8,8)):
    cm=confusion_matrix(y, y_pred)

    fig,ax=plt.subplots(1,1,figsize=figsize)
    mp=ax.imshow(cm.T)
    fig.colorbar(mp)
    u = np.unique(y)
    ax.set_xticks(np.arange(len(u)))
    ax.set_xticklabels(u)

    ax.set_yticks(np.arange(len(u)))
    ax.set_yticklabels(u)


    ax.set_xticks(np.arange(-.5, len(u), 1), minor=True)
    ax.set_yticks(np.arange(-.5, len(u),1), minor=True)

    ax.grid(which='minor', color='k', linestyle='-', linewidth=4)
    ax.tick_params(labelsize=15, direction='inout', size=10, width=2)

    ax.set_ylabel('Real label', fontsize=20)
    ax.set_xlabel('Predicted label', fontsize=20)

    for i in np.arange(0,len(u),1):
        for j in np.arange(0,len(u),1):
            ax.text(i,j,str(cm[i,j]),ha="center", va="center", color="k", fontsize=20)
    plt.show()
```

```python
def mini_branch_logistic_regression_mode(X_train,y_train,X_test,y_test,acc_stop=0.98,r=0.2,alpha=0.0001,N=100):
    """
    acc_stop - minimum accuracy
    r - mini branch size (rate)
    alpha - learning rate
    N - maximum iteration number
    
    references:
    https://agustinus.kristia.de/techblog/2016/06/21/nn-sgd/
    https://towardsdatascience.com/batch-mini-batch-stochastic-gradient-descent-7a62ecba642a
    https://machinelearningmastery.com/gentle-introduction-mini-batch-gradient-descent-configure-batch-size/
    """
    X=X_train
    y=y_train
    m,n=X.shape
    index = np.arange(m)

    w = np.random.uniform(0,1,n) # starting values
    b = np.random.uniform(0,10)
    L = []
    acc = []
    for i in range(N):
        random_index = np.arange(m)
        np.random.shuffle(random_index)
        indexes = np.split(random_index,np.array([int(i*m) for i in np.arange(r,1,r)]), axis=0)
        
        for index in indexes:
            X_train_mini = X[index]
            y_train_mini = y[index]
            
            # calculate the gradients
            y_pred = predict(X_train_mini,w,b)
            db = logistic_loss_db(y_pred, y_train_mini) 
            dw = logistic_loss_dw(X_train_mini,y_pred,y_train_mini)
            y_pred = predict(X_test,w,b)
            
            #update the model
            b = b - alpha * db
            w = w - alpha * dw

        y_pred = predict(X_test,w,b)
        #caculate loss and accuracy
        loss = np.sum(logistic_loss(y_pred, y_test)) / y_test.size
        L.append(loss)
        acc.append(ACC(y_pred,y_test))
        
        if acc[-1]>acc_stop: break
    
    #plot loss and accuracy as funtion of number of steps
    fig,ax=plt.subplots(1,1,figsize=(10,6))
    ax.set_title('Cost', fontsize=30)
    ax.plot(np.arange(len(L))+1,L, '-', color='tab:blue')
    ax.plot(np.arange(len(L))[::5]+1,L[::5], 'o', color='tab:blue')
    ax.tick_params(labelsize=20, size=15, width=2, direction='inout')
    ax.set_xlabel('iteration number', fontsize=20)
    ax.set_ylabel('cost', fontsize=20)
    ax.grid()
    plt.show()
    
    fig,ax=plt.subplots(1,1,figsize=(10,6))
    ax.set_title('Accuracy score', fontsize=30)
    ax.plot(np.arange(len(acc))+1,acc, '-', color='tab:blue')
    ax.plot(np.arange(len(acc))[::5]+1,acc[::5], 'o', color='tab:blue')
    ax.tick_params(labelsize=20, size=15, width=2, direction='inout')
    ax.set_xlabel('iteration number', fontsize=20)
    ax.set_ylabel('ACC', fontsize=20)
    ax.grid()
    plt.show()
    
    return w,b
```

```python
#load the provided data files
X_train=np.load('X_train.npy')
X_test=np.load('X_test.npy')

y_train=np.load('y_train.npy')
y_test=np.load('y_test.npy')

#b=np.load('b.npy')
#w=np.load('w.npy')
```

```python
#optimize the model
w,b = mini_branch_logistic_regression_mode(X_train, y_train, X_test, y_test)
```

```python
#create prediction and plot the confusion matrix
y_pred=predict(X_test,w,b)

conf_matrix(y_test, np.round(y_pred))
```

### 2. MLP model in `tf.keras`

* implement a model as such:

 * 750 neuron, relu activation
 * 500 neuron, relu activation
 * 500 neuron, relu activation
 * 10 neuron, softmax activation

* use the `tf.keras.datasets.mnist` dataset in this case
* preprocess if necessary
* use the 'sgd' optimizer, log the loss ('categorical_crossentropy') and the accuracy, as well as the AUC score during training, visualize the history of these values


```python
import tensorflow as tf
from tensorflow import keras
```

```python
#load and preporcess the data set
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

x_train = x_train.reshape(x_train.shape[0], x_train.shape[1]*x_train.shape[2]) / 255.
x_test  = x_test.reshape(x_test.shape[0], x_test.shape[1]*x_test.shape[2]) / 255.

y_train = keras.utils.to_categorical(y_train)
y_test  = keras.utils.to_categorical(y_test)
```

```python
#creating the model, references:
#https://patbaa.web.elte.hu/physdm/code_examples/fully_connected.html
#https://www.tensorflow.org/datasets/keras_example

model = keras.models.Sequential([keras.layers.InputLayer(input_shape=x_train[0].shape),
                                 keras.layers.Dense(750, activation='relu'),
                                 keras.layers.Dense(500, activation='relu'),
                                 keras.layers.Dense(500, activation='relu'),
                                 keras.layers.Dense(10, activation='softmax')])


model.compile(optimizer=keras.optimizers.SGD(),
              loss = 'categorical_crossentropy',
              metrics=['accuracy','AUC'],
)

model.summary()
```

```python
#fit the model
history = model.fit(x=x_train, y=y_train, epochs=15, validation_data=(x_test, y_test))
```

```python
fig,axes = plt.subplots(3,1, figsize=(10,24))

ax=axes[0]
ax.plot(history.history['loss'], label='train loss', ls='-', marker='o', color='tab:red')
ax.plot(history.history['val_loss'], label='val loss',ls='--', marker='s', color='tab:blue')
ax.set_ylabel('loss', fontsize=20)
ax.set_xlabel('epochs', fontsize=20)
ax.tick_params(labelsize=15,size=10,width=2,direction='inout')
ax.grid()
ax.set_title('Loss function',fontsize=25)
ax.legend(fontsize=20)

ax=axes[1]
ax.plot(history.history['accuracy'], label='train accuracy', ls='-', marker='o', color='tab:red')
ax.plot(history.history['val_accuracy'], label='val accuracy',ls='--', marker='s', color='tab:blue')

ax.set_ylabel('accuracy', fontsize=20)
ax.set_xlabel('epochs', fontsize=20)
ax.set_title('accuracy score',fontsize=25)
ax.legend(fontsize=20)
ax.tick_params(labelsize=15,size=10,width=2,direction='inout')
ax.grid()


ax=axes[2]
ax.plot(history.history['auc'], label='train auc', ls='-', marker='o', color='tab:red')
ax.plot(history.history['val_auc'], label='val auc',ls='--', marker='s', color='tab:blue')

ax.set_ylabel('AUC', fontsize=20)
ax.set_xlabel('epochs', fontsize=20)
ax.set_title('AUC', fontsize=25)
ax.legend(fontsize=20)
ax.tick_params(labelsize=15,size=10,width=2,direction='inout')
ax.grid()


plt.show()
```

```python
#my simple model

model = keras.models.Sequential([keras.layers.InputLayer(input_shape=x_train[0].shape),
                                 keras.layers.Dense(256, activation='relu'),
                                 keras.layers.Dense(256, activation='relu'),
                                 keras.layers.Dense(128, activation='relu'),
                                 keras.layers.Dense(128, activation='relu'),
                                 keras.layers.Dense(64, activation='relu'),
                                 keras.layers.Dense(64, activation='relu'),
                                 keras.layers.Dense(32, activation='relu'),
                                 keras.layers.Dense(32, activation='relu'),
                                 keras.layers.Dense(10, activation='softmax')])


model.compile(optimizer=keras.optimizers.SGD(),
              loss = 'categorical_crossentropy',
              metrics=['accuracy'],
)

model.summary()
```
